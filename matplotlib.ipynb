```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import style

style.use('fivethirtyeight')

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

def animate(i):
    graph_data = open('example.txt','r').read()
    lines = graph_data.split('\n')
    xs = []
    ys = []
    for line in lines:
        if len(line) > 1:
            x, y = line.split(',')
            xs.append(float(x))
            ys.append(float(y))
    ax1.clear()
    ax1.plot(xs, ys)
    
ani = animation.FuncAnimation(fig, animate, interval=1000)
plt.show()
```

## Animations in Matplotlib can be made by using the Animation class in two ways:

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import urllib.request as urq
from matplotlib import style
import matplotlib.ticker as mticker
from mplfinance.original_flavor import candlestick_ohlc

import matplotlib.dates as mdates
import requests

style.use('fivethirtyeight')

MA1 = 10
MA2 = 60

def moving_average(values, window):
    weights = np.repeat(1.0, window)/window
    smas = np.convolve(values, weights, 'valid')
    return smas

def high_minus_low(highs, lows):
    return highs-lows

def bytespdate2num(fmt, encoding='utf-8'):
    def bytesconverter(b):
        s = b.decode(encoding)
        return mdates.datestr2num(s)
    return bytesconverter

def graph_data():

    fig = plt.figure()
    ax1 = plt.subplot2grid((6,1), (0,0), rowspan=1, colspan=1)
    plt.title("Stock Price")
    ax2 = plt.subplot2grid((6,1), (1,0), rowspan=4, colspan=1)
    plt.xlabel('Date')
    plt.ylabel('Price')
    ax3 = plt.subplot2grid((6,1), (5,0), rowspan=1, colspan=1)
    
    stock_price_url='https://pythonprogramming.net/yahoo_finance_replacement'
    source_code = urq.urlopen(stock_price_url).read().decode()

    stock_data = []
    split_source = source_code.split('\n')

    for line in split_source:
        split_line = line.split(',')
        if len(split_line) == 7:
            if 'Volume' not in line:
                stock_data.append(line)

    datep, openp, highp, lowp, closep, ad_closep, vol = np.loadtxt(stock_data,
                                                                  delimiter=',',
                                                                  unpack=True,
                                                                  converters={0:bytespdate2num('%Y-%m-%d')}
                                                                  )
    
    ma1 = moving_average(closep,MA1)
    ma2 = moving_average(closep,MA2)
    start = len(datep[MA2-1:])

    h_l = list(map(high_minus_low, highp, lowp))

    ax1.plot_date(datep,h_l,'-')
    
    ohlc = list(zip(datep, openp, highp, lowp, closep, ad_closep, vol))
    candlestick_ohlc(ax2, ohlc, width=0.4, colorup='#77d879', colordown='#db3f3f')
    
    for label in ax2.xaxis.get_ticklabels():
        label.set_rotation(45)
    ax2.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    ax2.xaxis.set_major_locator(mticker.MaxNLocator(10))
    ax2.grid(True)#, color='g', linestyle='-', linewidth=5)
    
    ax2.annotate(str(closep[0]), (datep[0], closep[0]),
                 xytext = (datep[0]+300, closep[0]), bbox=dict(boxstyle='larrow',fc='w', ec='k',lw=1))
    
    ax3.plot(datep[-start:], ma1[-start:])
    ax3.plot(datep[-start:], ma2[-start:])
    
    plt.subplots_adjust(left=0.09, bottom=0.20, right=0.87, top=0.90, wspace=0.2, hspace=0)
    plt.show()

graph_data()
```
